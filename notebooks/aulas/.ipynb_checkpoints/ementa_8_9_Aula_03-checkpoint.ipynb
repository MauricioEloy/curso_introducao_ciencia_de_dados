<h1 style="background-color:#084d6e;font-family:newtimeroman;font-size:300%;color:white;text-align:center;border-radius: 15px 50px">Expressões Booleanas</h1>

Uma expressão booleada é aquela que pode receber duas opções ```True``` ou ```False```.

Para usarmos uma expressão booleana necessitamos de alguns operadores:

<a id="3.1"></a>
<h2 style="background-color:#084d6e;font-family:newtimeroman;font-size:200%;color:white;text-align:center;border-radius: 15px 50px">Operadores Relacionais</h2>

Operador | Significado | Exemplo
:-------:|:----------:|:-------:
== | é igual a | x==y
!= | é diferente de | x!=y
> | é maior que | x > y
< | é menor que | x < y
>= | é maior ou igual a | x >= y
<= | é menor ou igual a | x <=y


````python
x = 2
y = 3
z = 3

print("{}=={}: {}".format(x,y,x==y))
````

<a id="3.2"></a>
<h2 style="background-color:#084d6e;font-family:newtimeroman;font-size:200%;color:white;text-align:center;border-radius: 15px 50px">Operadores Lógicos</h2>

Operador | Significado
:-------:|:----------:
```and``` | uma expressão booleana é verdadeira se ambas as afirmações forem verdadeiras
```or``` | uma expressão booleana é verdadeira se uma das afirmações for verdadeira
```not``` | é usado para negar uma expressão booleana


````python
print("{} != {} e {} > {}---{}".format(x,y,x,y,(x!=y) and (x>y)))
````

````python
print("{} != {} ou {} > {}---{}".format(x,y,x,y,(x!=y) or (x>y)))
````

````python
not(x>y)
````

<a id="4.1"></a>
<h2 style="background-color:#084d6e;font-family:newtimeroman;font-size:200%;color:white;text-align:center;border-radius: 15px 50px">A Indentação</h2>

Cada linguagem de programação tem sua particularidade quanto a formação de um bloco de código.

Na linguagem ```Python``` o que denota os blocos de código é a **indentação**, deste modo, tudo que estiver em um mesmo nível de indentação pertence ao mesmo bloco.

A identação pode ser feita utilizando a tecla ```tab``` ou o se equivalente ```4 espaços```.

<a id="4.2"></a>
<h2 style="background-color:#084d6e;font-family:newtimeroman;font-size:200%;color:white;text-align:center;border-radius: 15px 50px">O Comando</h2>

Quando estamos escrevendo um programa, normalmente precisamos controlar o seu fluxo de execução por meio de de condições.

Este é o papel do comando **if** que verifica as condições e controla o fluxo do programa perante a resposta a resposta obtida (```True``` ou ```False```).

A estrutura básica do comando ```if``` é:

```python
if condicao_a_ser_analisada:
    comandos_executados_caso_a_resposta_seja_True
```

OBS: Para um controle com mais comparações, pode-se combinar o comando ```if``` com os comandos ```else``` (senão) e ```elif``` (senão se):

<a id="4.3"></a>
<h2 style="background-color:#084d6e;font-family:newtimeroman;font-size:200%;color:white;text-align:center;border-radius: 15px 50px">Exemplos</h2>



````python
x = float(input("x= "))
y = float(input("y= "))
z = float(input("z= "))
````

````python
if x==y:
    print("x é igual à y")
````

````python
if x==y:
    print("x é igual à y")
else:
    print("x é diferente de y") 
````

````python
print("x é igual à y" if x==y else "x é diferente de y")
````

````python
if x==y and y==z:
    print("Todos os números são iguais")
elif x==y or y==z:
    print("Temos dois números iguais")
else:
    print("Todos os números são diferentes")
````

<h1 style="background-color:#084d6e;font-family:newtimeroman;font-size:300%;color:white;text-align:center;border-radius: 15px 50px">Iteração</h1>

**Iterar** é a execução repetida de uma sequência de instruções.

Esse tipo de tarefa é realizado com maestria pelos computadores, uma vez que eles o fazem sem cometer erros,o que não podemos dizer para as pessoas.

Vamos ver dois tipos de iteradores:
1. O comando ```for```;
2. O comando ```while```.

<a id="5.1"></a>
<h2 style="background-color:#084d6e;font-family:newtimeroman;font-size:200%;color:white;text-align:center;border-radius: 15px 50px">Comando "for"</h2>

O camando ```for``` em Python é um iterador que pode ser usado para iterar em qualquer tipo de sequência.

A estrutura básica do comando ```for``` é:

```python
for valor in sequencia:
    bloco_de_codigo_executado
```

````python
num_primos_menores_10 = (2,3,5,7)

for p in num_primos_menores_10:
    print(p)
````

````python
lista_padaria = ["pao", "leite", "cafe"]

for item in lista_padaria:
    print(item)
````

````python
frase = "Já chegou!!!"

for coisa in frase:
    print(coisa)
````

````python
dic_reuniao = {'pizza': 10,
              'refrigerante': 5,
              'suco': 3,
              'chocolate': 5}

for chave, valor in dic_reuniao.items():
    print("Preciso comprar {} {}".format(valor,chave))
````

````python
dic_reuniao.items()
````

````python
for par in dic_reuniao.items():
    print("Preciso comprar {} {}".format(par[1],par[0]))
````

<a id="5.2"></a>
<h2 style="background-color:#084d6e;font-family:newtimeroman;font-size:200%;color:white;text-align:center;border-radius: 15px 50px">A função range()</h2>

É uma função utilizada para criar sequências de inteiros e pode assumir um, dois ou três argumentos.

```python
range(inicio, fim, passo)
```

````python
for n in range(10):
    print(n)
````

````python
for n in range(1,11):
    print(n)
````

````python
for n in range(0,11,2):
    print(n)
````

<a id="5.3"></a>
<h2 style="background-color:#084d6e;font-family:newtimeroman;font-size:200%;color:white;text-align:center;border-radius: 15px 50px">Comando "while"</h2>

O camando ```while``` em Python é uma outra maneira de fazermos iterações.

A estrutura básica do comando ```while``` é:

```python
while teste:
    bloco_de_codigo_executado
```

O ```bloco_de_codigo_executado``` será executado enquanto o valor do ```teste``` for verdadeiro.

Podemos usar uma estrutura de ```while``` associada ao comando ```else```:

```python
while teste:
    bloco_de_codigo_executado_1
else:
    bloco_de_codigo_executado_2
```

````python
cont = 0

while cont <=10:
    print(cont)
    cont = cont + 1    
````

````python
cont = 0

while cont <=10:
    print(cont)
    cont += 1
else:
    print("FIM")
````

<a id="5.4"></a>
<h2 style="background-color:#084d6e;font-family:newtimeroman;font-size:200%;color:white;text-align:center;border-radius: 15px 50px">Intruções break, continue e pass</h2>

Quando usamos iteradores, podemos nos deparar com fatores externos que podem influenciar de alguma maneira o nosso programa, deste modo, podemos usar algums instruções para realizarmos algumas instruções como:

* sair do loop completamente - ```break```
* ignorar parte de um loop antes de continuar - ```continue```
* ignorar o fator externo - ```pass```

````python
for num in range(1,11):
    if num == 5:
        break
        
    print("O número é " + str(num))
    
print("Fim do loop")
````

````python
for num in range(1,11):
    if num == 5:
        continue
        
    print("O número é " + str(num))
    
print("Fim do loop")
````

````python
for num in range(1,11):
    if num == 5:
        pass
        
    print("O número é " + str(num))
    
print("Fim do loop")
````
