<h1 style="background-color:#084d6e;font-family:newtimeroman;font-size:300%;color:white;text-align:center;border-radius: 15px 50px">Módulos e Pacotes</h1>

Módulos são simplesmente arquivos Python (extensão .py), que implementam um conjunto de funções e podem ser importados.

Sintaxe:
```python
import nome_modulo
```

Um pacote é um módulo com um conjunto de módulos.

Sintaxe:

```python
import nome_pacote.nome_modulo
```

<a id="2.1"></a>
<h2 style="background-color:#084d6e;font-family:newtimeroman;font-size:200%;color:white;text-align:center;border-radius: 15px 50px">Comando Import</h2>



````python
def media(ls_numbers):
    media = sum(ls_numbers)/len(ls_numbers)
    return media
````

````python
vel_iot = [60.0, 55.7, 43.0, 79.3, 41.4, 14.7]
media(vel_iot)
````

````python
import numpy

numpy.mean(vel_iot)
````

````python
mean(vel_iot)
````

````python
import numpy as np

np.mean(vel_iot)
````

<h2 style="background-color:#084d6e;font-family:newtimeroman;font-size:200%;color:white;text-align:center;border-radius: 15px 50px">Atributos e Métodos</h2>

Um módulo pode ter atributos e métodos definidos, a sintaxe para acessá-los é a mesma ```nome_modulo.nome_metodo``` ou ```nome_modulo.nome_atributo```.

Para verificar os atributos e métodos existentes em um módulo pode-se:

1. utilizar do método ```dir```;
2. acessar a documentação do módulo/pacote.


````python
dir(np)
````

Duas formas de obter a documentação:
1. https://pypi.org/;
2. pesquisa.

<h2 style="background-color:#084d6e;font-family:newtimeroman;font-size:200%;color:white;text-align:center;border-radius: 15px 50px">Comando from</h2>

Através do comando ```from``` pode-se importar apenas elementos específicos de um módulo/pacote.

Sintaxe:

```python
from nome_modulo import nome_especifico
```

````python
from numpy import mean
````

````python
mean(vel_iot)
````
