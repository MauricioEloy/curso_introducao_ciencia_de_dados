<h1 style="background-color:#084d6e;font-family:newtimeroman;font-size:300%;color:white;text-align:center;border-radius: 15px 50px">Comprehension</h1>

É uma maneira concisa e eficaz de construir uma sequência, onde cada elemento é obtido através do resultado da aplicação de uma expressão a cada item de outra sequência ou iterável.

<a id="2.1"></a>
<h2 style="background-color:#084d6e;font-family:newtimeroman;font-size:200%;color:white;text-align:center;border-radius: 15px 50px">Geral</h2>



````python
vel_iot = [60.0, 55.7, 43.0, 79.3, 41.4]

list_for = []

for vel in vel_iot:
    vel_ms = round(vel/3.6,2)
    list_for.append(vel_ms)
    
list_for
````

````python
list_comp = [round(vel/3.6,2) for vel in vel_iot]

list_comp
````

Estrutura básica do comprehension:

```python
list_comp = [expressao for variavel in objeto]
```

Como vimos, o resultado do comprehension acima é o mesmo que se tivéssemos utilizado o comando "for":

```python
list_comp = []
for variavel in objeto:
    list_comp.apprend(expressao)
```

Além da escrita, qual a vantagem do uso do comprehension?

````python
import random

list_vel_iot = [random.randrange(0,80) for i in range(1000000)]

list_vel_iot
````

````python
%%timeit   

#utilizando "for"

vel_iot_ms_for = []

for vel in list_vel_iot:
    vel_ms = round(vel/3.6, 2)
    vel_iot_ms_for.append(vel_ms)
````

````python
%%timeit

#utilizando comprehension

vel_iot_ms_comp = [round(vel/3.6, 2) for vel in list_vel_iot]
````

<h2 style="background-color:#084d6e;font-family:newtimeroman;font-size:200%;color:white;text-align:center;border-radius: 15px 50px">Combinado</h2>

Podemos ter várias maneiras de estruturar um código, muitas delas podem combinar um comando ```for``` com o condicional ```if```, como no exemplo abaixo:

```python
list_comp = []
for variavel in objeto:
    if condicao:
        list_comp.apprend(expressao)
```

De maneira equivalente, podemos ter um comprehension:
```python
list_comp = [expressao for variavel in objeto if condicao]
```

````python
vel_iot = [random.randrange(0,80) for i in range(100)]

vel_iot_ms_comp_1 = [round(vel/3.6, 2) for vel in vel_iot if vel>=15]

len(vel_iot) - len(vel_iot_ms_comp_1)
````

````python
vel_iot_ms_comp_2 = [round(vel/3.6, 2) if vel>=15 else "<15" \
                     for vel in vel_iot]

len(vel_iot) - len(vel_iot_ms_comp_2)
````

````python
vel_iot_ms_comp_2.count("<15")
````

<h1 style="background-color:#084d6e;font-family:newtimeroman;font-size:300%;color:white;text-align:center;border-radius: 15px 50px">Funções</h1>

Suponha agora que você tenha que executar a tarefa anterior por diversas vezes durante o seu código e com valores variados para o filtro
```python 
if vel>=15
```

A solução para este problema está no uso do que chamamos de **Função**. 

Em programação, uma função é uma sequência de comandos que executa uma determinada tarefa. As funções recebem um nome e também nos permitem especificar os parâmetros que serão utilizados.

Para criarmos uma função devemos utilizar do comando ```def```, ao utilizarmos este comando estamos criando um objeto do tipo função.

A sintaxe de definição de uma função é:
```python
def nome_da_funcao(parametros):
    estrutura_de_codigo
```

````python
def convertVel(list_vel_iot,limit):
    vel_iot_ms = [round(vel/3.6, 2) if vel>=limit else "<"+str(limit)\
              for vel in list_vel_iot]
    print("Foram encontradas {} velocidades menores\
        que {} km/h".format(vel_iot_ms.count("<"+str(limit)),limit))
    return vel_iot_ms
````

<h2 style="background-color:#084d6e;font-family:newtimeroman;font-size:200%;color:white;text-align:center;border-radius: 15px 50px">Definição de Parâmetros</h2>

Os parâmetros/argumentos de uma função pode ser fornecidos na criação da mesma, de modo que se não definidos irão ser utilizados os valores pré-estabelecidos (default).

````python
def convertVel(list_vel_iot,limit,casas=2):
    list_iot_ms = [round(vel/3.6, casas) if vel>=limit \
                   else "<"+str(limit) for vel in list_vel_iot]
    print("Foram encontradas {} velocidades menores\
        que {} km/h".format(list_iot_ms.count("<"+str(limit)),limit))
    return list_iot_ms
````

````python
convertVel(vel_iot, 15)
````

````python
convertVel(vel_iot, 15,3)
````

<h2 style="background-color:#084d6e;font-family:newtimeroman;font-size:200%;color:white;text-align:center;border-radius: 15px 50px">Funções Lambda</h2>

A função lambda permite a criação de funções "anônimas" em tempo de execução, sem a necessidade de a definirmos anteriormente através do uso do ```def```.

Este tipo de expressão é muito útil em estruturas de código em que não se pode utilizar o uso do ```def```, como por exemplo, em comprehension.

A sua sintaxe é:
```python
lambda argumentos: expressao
```

````python
f_x = lambda x: x**2 + 2*x
f_x(1)
````

````python
f_xy = lambda x,y: x**2 + y**2
f_xy(1,1)
````
